```python
import pandas as pd
import matplotlib.pyplot as plt

# results.csv   
results_file = 'results.csv'

# results.csv  
df = pd.read_csv(results_file)

# Loss  
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(df['               epoch'], df['      train/obj_loss'], label='Objectness Loss')
ax.plot(df['               epoch'], df['      train/box_loss'], label='Box Loss')
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
ax.set_title('Training Losses')
ax.legend()
plt.tight_layout()
plt.show()

# mAP  
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(df['               epoch'], df['     metrics/mAP_0.5'], label='mAP@0.5')
ax.plot(df['               epoch'], df['metrics/mAP_0.5:0.95'], label='mAP@0.5:0.95')
ax.set_xlabel('Epochs')
ax.set_ylabel('mAP')
ax.set_title('mAP Scores')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

def find_best_epoch(results_file):
    df = pd.read_csv(results_file)
    # 'epoch'    
    best_epoch = df['               epoch'][df['metrics/mAP_0.5:0.95'].idxmax()]
    results = df.to_dict('records')
    return best_epoch, results

results_file = 'results.csv'  #    
best_epoch, results = find_best_epoch(results_file)

print(f"Best Epoch: {best_epoch}")
```
